```python
# import library
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

```python
#buat array 2D ke dalam variabel x. data ini yang akan diklaster
x = np.array([[1,2],[2.5,4.5],[2,2],[4,1.5],[4,2.5]])
```

```python
print (x)
```

```python
#lakukan hieararchical klastering
z = linkage(x)
```

```python
print (z)
```

```python
#plot dendrogram
plt.figure(figsize=(25, 10))
plt.title('Dendrogram Hieararchical Klastering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(z)
plt.show()
```
